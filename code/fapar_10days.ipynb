```python
Y1=2018
Y2=2024

fapar_all_list = []
for Y in range(Y1,Y2+1):
    fapar_year_list = []
    for m in range(1,13):
        try:
           print('/Volumes/blue_wd/ESA_F4R/fapar/'+str(Y)+'/c3s_FAPAR_'+str(Y)+"{:02d}".format(m)+'*_GLOBE_SENTINEL3_V4.1.0.area-subset.12.32.-15.3.nc')
           fapar = xr.open_mfdataset('/Volumes/blue_wd/ESA_F4R/fapar/'+str(Y)+'/c3s_FAPAR_'+str(Y)+"{:02d}".format(m)+'*_GLOBE_SENTINEL3_V4.1.0.area-subset.12.32.-15.3.nc')['fAPAR']
           fapar = fapar.sel(lat=slice(12,5),lon=slice(10,31),drop=True).load()
           fapar = fapar.sortby('lat', ascending=True)
           fapar_year_list.append(fapar)
           del fapar
           gc.collect()
        except:
            print('no month - ',m,' for year - ',Y)
    fapar_year = xr.concat(fapar_year_list,dim='time')
    del fapar_year_list
    gc.collect()
    fapar_year.to_netcdf(dataf+'fapar_'+str(Y)+'_10day_n.nc',engine='h5netcdf')
    del fapar_year
    gc.collect()
    print('done - ',Y)
        
```
